```python
df = df[df['annual_income'] <= 800000000]
```

```python
df = df[df['age'] <= 100]
rfm_df = df[['recency','total_amount','total_count']]
rfm_df.describe()
```

[마지막으로 구매하고 지난 날 0~99 ] 
- 0 ~ 10일 경과 -> +1

[구매한 빈도 수 0~33]
- 5회 단위로 -> +1

[구매 금액 6,500 ~ 3,282,500]
- 50000원 단위로 -> +1

RFM 분석 방법 사용

```python
# 점수를 계산하는 함수
def calculate_recency_score(days):
    return 10 - (days // 10)

def calculate_frequency_score(count):
    return (count // 5)

def calculate_monetary_score(amount):
    return amount // 50000

# 데이터프레임의 각 행에 대해 총점을 계산하는 함수
def calculate_total_score(row):
    recency = calculate_recency_score(row['recency'])
    frequency = calculate_frequency_score(row['total_count'])
    monetary = calculate_monetary_score(row['total_amount'])
    return recency + frequency + monetary

# 새로운 열 'total_score' 추가
df['total_score'] = rfm_df.apply(calculate_total_score, axis=1)
print(df['total_score'])
```

```python
df.describe()
```

total_score 점수 1~75까지 있음
- 아이언 회원 1~ 10점
- 브론즈 회원 10 ~ 30점
- 실버 회원 31 ~ 50점
- 골드 회원 51 ~ 60점
- 다이아 회원 61 ~ 70점 
- 마스터 회원 71 점 이상

```python
def user_score (row):
    score = row['total_score']

    if score <= 15:
        return '아이언 회원'
    elif score <=30:
        return '브론즈 회원'
    elif score <= 45:
        return '실버 회원'
    elif score <= 60:
        return '골드 회원'
    else:
        return '다이아 회원'

df['membership_level'] = df.apply(user_score, axis=1)
df['membership_level']
```

### 멤버십 레벨별 어떤 프로모션을 많이 참여했는지

```python
promotion_df = df[['membership_level','promotion_1','promotion_2','promotion_3','promotion_4','promotion_5','promotion_6']]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 프로모션별 멤버십 레벨 참여 인원 집계
promotions_participation = pd.DataFrame()

# 멤버십 레벨 순서
membership_order = ['아이언 회원', '브론즈 회원', '실버 회원', '골드 회원', '다이아 회원']

for i in range(1, 7):  # 프로모션 1부터 6까지
    column_name = f'promotion_{i}'
    # 해당 프로모션에 참여한 사람들만 걸러내기
    participated = promotion_df[promotion_df[column_name] == 1]
    # 멤버십 레벨별 참여 인원 집계
    counts = participated['membership_level'].value_counts().reindex(membership_order, fill_value=0)
    promotions_participation[column_name] = counts

# NaN 값을 0으로 채우기 (참여 인원이 없는 레벨)
promotions_participation.fillna(0, inplace=True)

# 시각화
plt.figure(figsize=(14, 8))
sns.heatmap(promotions_participation, annot=True, fmt=".0f", cmap="YlGnBu", linewidths=.5)
plt.title('멤버십 레벨별 프로모션 참여 인원')
plt.xlabel('프로모션')
plt.ylabel('멤버십 레벨')
plt.yticks(rotation=0)  # y축 레이블 회전 없음
plt.show()
```

- 전체적으로 브론즈~골드 회원들의 프로모션 참여률이 대체적으로 높다
- 다이아~마스터 회원들은 프로모션 참여에 크게 관심이 없다

- promotion2는 사람들이 많이 참여를 안했다. 
- promotion6은 사람들이 가장 많이 참여를 했다.
